# Euler method

```python
import numpy as np
import matplotlib.pyplot as plt

# Plotting parameters:
csfont = {'fontname':'Charter', 'fontweight':'regular'}
hfont = {'fontname':'Charter', 'fontweight':'bold'}
ifont = {'fontname':'Charter', 'fontweight':'regular', 'style':'italic'}

font_axes = 12
font_labels = 18
font_title = 22
font_text = 14
font_legend = 16
```

***

## Example 1

This example approximates the solution to the DE:

$df(x)/dx = e^x$

subject to:

$f(x_{bc}) = e^{x_{bc}}$

with the Euler method.

The analytical solution $f(x) = e^x$ is plotted with the approximation.

```python
# Boundary condition:
x_bc = 1
f_bc = np.exp(x_bc)

# Number of Euler steps:
N_euler = 6

# Discretization of x-axis for the Euler method approximation:
x_end = 4
x_step = (x_end - x_bc)/N_euler
x_step_euler = (x_end - x_bc)/N_euler
x_euler = np.linspace(x_bc, x_end, N_euler+1)

# Initialize approximation function:
F_euler = np.zeros(N_euler+1)

# Set boundary condition:
F_euler[0] = f_bc

for i in range(0,N_euler):
    F_euler[i+1] = F_euler[i] + x_step_euler * np.exp(x_euler[i])

# Analytical solution:
N = 30
x_real = np.linspace(x_bc, x_end, N)
F_real = np.exp(x_real)

# Plot:
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the analytical solution:
plt.plot(x_real,F_real, 'b-')

# Plot the Euler method approximation:
plt.plot(x_euler,F_euler, 'ro--')

plt.grid(True, alpha=0.1)
plt.xlabel(r'x', fontsize=font_labels, **csfont)
plt.ylabel(r'f(x)', fontsize=font_labels, **csfont)
lgnd = plt.legend(['Original', 'Approx'], fontsize=font_legend, loc="upper left")

# Save plot:
filename = 'euler-method-e-to-x.png'
plt.savefig(filename, dpi = 150)
```

## Example 2

This example approximates the solution to the DE:

$df(x)/dx = sin(x)$

subject to:

$f(x_{bc}) = -cos(x_{bc})$

with the Euler method.

The analytical solution $f(x) = -cos(x)$ is plotted with the approximation.

```python
# Boundary condition:
x_bc = np.pi
f_bc = -np.cos(x_bc)

# Number of Euler steps:
N_euler = 5

# Discretization of x-axis for the Euler method approximation:
x_end = 6*np.pi
x_step = (x_end - x_bc)/N_euler
x_step_euler = (x_end - x_bc)/N_euler
x_euler = np.linspace(x_bc, x_end, N_euler+1)

# Initialize approximation function:
F_euler = np.zeros(N_euler+1)

# Set boundary condition:
F_euler[0] = f_bc

for i in range(0,N_euler):
    F_euler[i+1] = F_euler[i] + x_step_euler * np.sin(x_euler[i])

# Analytical solution:
N = 100
x_real = np.linspace(x_bc, x_end, N)
F_real = -np.cos(x_real)

# Plot:
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the analytical solution:
plt.plot(x_real,F_real, 'b-')

# Plot the Euler method approximation:
plt.plot(x_euler,F_euler, 'ro--')

plt.grid(True, alpha=0.1)
plt.xlabel(r'x', fontsize=font_labels, **csfont)
plt.ylabel(r'f(x)', fontsize=font_labels, **csfont)
lgnd = plt.legend(['Original', 'Approx'], fontsize=font_legend, loc="lower left")
    
# Save plot:
filename = 'euler-method-sin-unfortunate.png'
plt.savefig(filename, dpi = 150)
```
